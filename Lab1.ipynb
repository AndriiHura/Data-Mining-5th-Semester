# Сумма значений всех аттрибутов

```python
#Make dict with sums of all values
def sum_list_func(df_list):
    """Counts sums of all attributes"""
    sum_list = []
    for df in df_list:
        sum_dict = {}
        for att in attr_values:
            value_sum = 0
            for value in df[att].values:
                value_sum += value
            sum_dict[att] = value_sum
        sum_list.append(sum_dict)
   # print(sum_list)
    #print(df_list[0].count()[1])
    return sum_list
sum_list = sum_list_func(df_list)
print(sum_list_func(df_list))
```

# Математическое ожидание

```python
#Dict with average expected value
def av_ex_val(df_list):
    average_expected_value_list = []
    iterrator = 0
    for sum_dict in sum_list:
        average_expected_value_dict = {}

        for att in attr_values:
            average_expected_value = sum_dict[att]/df_list[iterrator].count()[1]
            average_expected_value_dict[att] = average_expected_value
        iterrator += 1
        average_expected_value_list.append(average_expected_value_dict)
    #print(average_expected_value_list)
    return average_expected_value_list

average_expected_value_list = av_ex_val(df_list)
print(av_ex_val(df_list))
```
